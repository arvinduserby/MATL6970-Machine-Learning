```python
# import required packages
import pandas as pd
import numpy as np
```

1) Import data from session storage (temporary, applies only to this notebook)

```python
# Load "Failure data.xlsx" save as failure_data
colnames = ['RF_temp','DC_temp']
failure_data = pd.read_excel("Failure data.xlsx", names=colnames)
```

```python
RF_temp1 = failure_data["RF_temp"].values
RF = RF_temp1[~np.isnan(RF_temp1)]
print(RF)
```

```python
DC = failure_data["DC_temp"].values
print(DC)
```

```python
Weibull_probability_plot(failures=RF)
plt.xlabel('Cycles to failure')
plt.ylabel('Occurence CDF%')
plt.title("Thermal cycling reliability \n RF data \n Maximum Likelihood Estimation (MLE)")
plt.show()
```

```python
Weibull_probability_plot(failures=DC)
plt.xlabel('Cycles to failure')
plt.ylabel('Occurence CDF%')
plt.title("Thermal cycling reliability \n DC data \n Maximum Likelihood Estimation (MLE)")
plt.show()
```

```python
Weibull_probability_plot(failures=RF, color = 'r')
Weibull_probability_plot(failures=DC, color ='g')
plt.xlabel('Cycles to failure')
plt.ylabel('Occurence CDF%')
plt.title("Thermal cycling reliability \n RF vs DC data \n Maximum Likelihood Estimation (MLE)")
# To show the legend correctly, we need to replot some points in separate scatter plots to create different legend entries
x, y = plotting_positions(failures=RF)
x, y = plotting_positions(failures=DC)
plt.scatter(x[0], y[0], color = 'r', label='RF')
plt.scatter(x[0], y[0], color = 'g', label='DC')
plt.legend()
plt.show()
```
